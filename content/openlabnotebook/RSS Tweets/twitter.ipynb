```python
import pandas as pd
import webbrowser
from tkinter import filedialog
```

```python
# download followers or followings by handle from 
url = 'https://www.twtdata.com/'
webbrowser.open(url)
```

```python
file  = filedialog.askopenfilename(title='Choose downloaded csv file:')
```

```python
# format twitter handles into nitter.net urls with /rss feed
df = pd.DataFrame({'text': pd.read_csv(file)['name'] + ' / ' + '@' + pd.read_csv(file)['username'],
                   'type': 'rss',
                   'xmlUrl': 'https://nitter.net/' + pd.read_csv(file)['username'] + '/rss',
                   'htmlUrl': 'https://nitter.net/' + pd.read_csv(file)['username'],
                   'description': 'Twitter feed for: ' + '@' + pd.read_csv(file)['username']})
df.set_index('text')
```

```python
# convert data frame to OPML XML
df.to_xml('/Users/guillermo/Downloads/out.opml.xml', row_name='outline', index=False, 
            attr_cols=['text', 'type', 'xmlUrl', 'htmlUrl', 'description'])

# how to indent correctly?

pd.read_xml('/Users/guillermo/Downloads/out.opml.xml')
```
